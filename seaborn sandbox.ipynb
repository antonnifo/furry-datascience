```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
# df = tips.copy()
# df.sample(7)
tips.head()
```

## Distribution plots

```python
# distplot 
sns.distplot(tips['total_bill']);
```

```python
# joint plot
sns.jointplot(x='total_bill',y='tip',data=tips);
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',color='green');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',color='green');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',color='green');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid',color='green');
```

```python
sns.pairplot(tips,hue='sex',palette='rainbow');
```

```python
sns.kdeplot(tips['total_bill']);
sns.rugplot(tips['total_bill']);
```

### categorical data

```python
sns.barplot(x='sex', y='total_bill',data=tips);
```

```python
sns.barplot(x='smoker',y='total_bill', data=tips,estimator=np.std);
```

```python
sns.countplot(x="sex",data=tips);
```

```python
sns.boxplot(x='day',y='total_bill', data=tips,hue='sex');
```

```python
sns.violinplot(x='day',y='total_bill',data=tips);
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker');
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True);
```
