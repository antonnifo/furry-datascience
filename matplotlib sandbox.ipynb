```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
z = x**3
```

```python
plt.plot(x,y,"g--")
plt.title("Title tu")
plt.xlabel("data")
plt.ylabel("x squared");
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"g*-")
plt.title("Squared")
plt.subplot(1,2,2)
plt.plot(x,z,"r--")
plt.title("Cubic")
```

```python
fig  = plt.figure(figsize=(12,6))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,color="yellow")
axes.set_title("Squared")
axes.set_xlabel("X")
axes.set_ylabel("X-squared");
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
ax = fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y,"g*-")
ax.plot(x,z,"yo-")
```

```python
fig1,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("Squared")
axes[1].plot(x,z)
axes[1].set_title("Cubic")
    
plt.tight_layout()    
```

```python
fig = plt.figure(figsize=(12,6))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,"g*-",label="Squred")
axes.plot(x,z,"y--",label="Cubic")

fig.legend(loc=1)
fig.savefig("figa.png")
```
