## Hierarchical Indxing

```python
import numpy as np
import pandas as pd
```

```python
# index levels
index = [("G1", 1),("G1", 2),("G1", 3),("G2", 1),("G2", 2),("G2", 3),("G3", 1),("G3", 2),("G3", 3)]
heir_index = pd.MultiIndex.from_tuples(index)
heir_index
```

```python
index2 = [("G1","G2","G3"),(1,2,3)]
heir_index2 = pd.MultiIndex.from_product(index2)
heir_index2
```

```python
#building the multi-dimensional dataframe
df = pd.DataFrame(np.random.rand(9,7),heir_index2,columns=["data1","data2","data3","data4","data5","data6","data7"])
df
```
