# Multi-Dimensional-Rows-&-Columns

```python
import numpy as np
import pandas as pd
```

```python
# Hierarichal rows and columns
columns = pd.MultiIndex.from_product([["Tony","Chris","Joe"],["HR","TEMP"]],names=["Subject","Type"])
index   = pd.MultiIndex.from_product([[2019,2018],[1,2]],names=["Year","Visit"])
```

```python
# Generate toy health data
data = np.random.rand(4,6)
data*=10
data = np.round(data)
data+=31
```

```python
#Build Multi-dimensional rows and columns
health_data = pd.DataFrame(data,index=index,columns=columns)
health_data
```

```python
health_data["Chris"]
```

```python
health_data["Chris","HR"]
```
